```python
from langdetect import detect #esta es con la funcion que llamamos arriba para detectar el idioma de un texto
#llamamos al corpus y filtramos por texto y que detecte el idioma y vea que es español
corpus = corpus[
    corpus["texto_limpio"].apply(
        lambda x: detect(str(x)) == "es"
    )
]
```

```python
import re #una libreria que tiene python para limpiar texto 
def limpiar_texto(texto):
    texto = str(texto).lower()

    # quitar urls
    texto = re.sub(r"http\S+|www\S+", "", texto)

    # quitar emojis
    texto = re.sub(r"[\U00010000-\U0010ffff]", "", texto)

    # quitar saltos de línea y tabs
    texto = re.sub(r"[\n\r\t]", " ", texto)

    # quitar caracteres que no sean letras españolas o espacios
    texto = re.sub(r"[^a-záéíóúñü\s]", "", texto)

    # quitar espacios repetidos
    texto = re.sub(r"\s+", " ", texto).strip()

    return texto
```

```python
corpus["texto_limpio"] = corpus["texto"].apply(limpiar_texto)
#quitamos vaores nulls 
corpus = corpus.dropna(subset=["texto_limpio"])
#quitamos texto donde queda solo
corpus = corpus[corpus["texto_limpio"] != ""]
#esto elimina el texto donde se 5 o menos caracteres
corpus = corpus[corpus["texto_limpio"].str.len() > 5]

#aqui comparamos cada uno de los textos uno limpio y el que teniamos para ver la diferencia 
corpus[["texto", "texto_limpio"]].head()
```

```python
corpus.to_csv(
    "resenas_clean.csv",
    index=False,
    encoding="utf-8-sig"
)
print()
```
